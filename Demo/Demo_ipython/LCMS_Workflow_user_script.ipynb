```python
import pandas as pd
import numpy as np

from corna.inputs import maven_parser as parser
from corna import constants as cons
from corna.helpers import merge_multiple_dfs
from corna.algorithms.nacorr_lcms import na_correction
from corna.postprocess import fractional_enrichment
```

```python
df= pd.read_csv('raw_intensity_file_LCMS.csv')
iso_tracers= ['C13', 'N15']
autodetect= False
ppm_input_user= 13
na_dict = {'C':[0.9889,0.0111],
'N':[0.9964,0.0036],
'O':[0.9976,0.0004,0.002],
'H':[0.99985,0.00015],
'S':[0.95,0.0076,0.0424],
}
eleme_corr = {}
```

```python
merged_df= parser.convert_inputdata_to_stdfrom(df)
```

```python
na_corrected, element_corr_dict =na_correction(merged_df, iso_tracers, ppm_input_user, na_dict, eleme_corr,autodetect=False)
print na_corrected
```

```python
fractional_enriched_df = fractional_enrichment(na_corrected)
print fractional_enriched_df
```

```python
merged_df = merge_multiple_dfs([na_corrected,fractional_enriched_df])
print merged_df
```
